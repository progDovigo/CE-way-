```python
renda = 5000.0
idade = 30
tempo_emprego = 6
score = 800
dividas = 'n'
valor_emprestimo = 10.000
limite = renda * 12 * 0.3

criterio_1 = renda > 3000 and (idade > 25 or tempo_emprego > 5)
print(criterio_1)

criterio_2 = score > 700 and dividas == "n"
print(criterio_2)

geral = valor_emprestimo < limite
print(geral)

aprovado = (criterio_1 or criterio_2) and geral
print(aprovado)
```
