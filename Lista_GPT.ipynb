Exercício 1 (Trivial)
Crie um programa que receba um número inteiro e verifique se ele é maior que 10. Exiba True ou False como resultado da expressão booleana.


```python
n = int(input("Digite um número inteiro: "))

if n > 10:
  print("True")
else:
  print("False")
```

Exercício 7
Crie um programa que receba um número e verifique se ele é múltiplo de 3 OU de 5, mas NÃO de ambos ao mesmo tempo.


```python
n = float(input("Digite um número: "))

if n % 3 == 0 and n % 5 != 0:
  print("True")
elif n % 5 == 0 and n % 3 != 0:
  print("True")
else:
  print("False")
```

Exercício 8
Receba três valores booleanos (True ou False) e determine se pelo menos dois deles são True. Resolva usando expressões booleanas.


```python
a = input("true ou false: ").lower()
b = input("true ou false: ").lower()
c = input("true ou false: ").lower()

if a == "true" and b == "true":
  print("True")
elif a == "true" and c == "true":
  print("True")
elif b == "true" and c == "true":
  print("True")
else:
  print("False")
```

Exercício 9 (Difícil)
Crie um programa que receba três números e verifique se eles podem formar um triângulo. Use condições booleanas considerando que a soma de dois lados deve ser maior que o terceiro.


```python
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))
c = float(input("Digite o terceiro número: "))

if a + b > c and a + c > b and b + c > a:
  print("pode formar um triângulo")
else:
  print("Não forma um triângulo")
```

Exercício 10 (Desafiador)
Receba três valores booleanos A, B e C. Construa uma expressão booleana equivalente à seguinte lógica:
"A expressão deve ser True somente quando exatamente dois dos três valores forem True OU quando todos forem False".
Implemente essa lógica usando operadores AND, OR e NOT.


```python
a = input("true or false: ").lower().strip()
b = input("true or false: ").lower().strip()
c = input("true or false: ").lower().strip()

if a == "true" and b == "true" and c == "false":
  print("TRUE")
elif a == "true" and b == "false" and c == "true":
  print("TRUE")
elif a == "false" and b == "true" and c == "true":
  print("TRUE")
elif a == "false" and b == "false" and c == "false":
  print("TRUE")
else:
  print("FALSE")
```
