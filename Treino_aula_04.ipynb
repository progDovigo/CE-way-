```python
ano = int(input("Insira o ano: "))

if ano % 4 == 0 and ano % 100 != 0 or ano % 400 == 0:
  print(f"o ano {ano} é bissexto!")
else:
  print(f"o ano {ano} não é bissexto!")
```

```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
op = input("Qual será a operação? (+, -, /, *)")

if op == "+":
  print(f"o resultado é: {num1 + num2}")
elif op == "-":
  print(f"o resultado é: {num1 - num2}")
elif op == "*":
  print(f"o resultado é: {num1 * num2}")
elif op == "/":
  if num2 == 0:
    print("Não podemos dividir por 0")
  else:
    print(f"o resultado é: {num1 / num2}")
else:
  print("Operador inválido. ")

```

```python
letra = input("Insira uma letra: ").lower()
if letra == "a" or letra == "e" or letra == "i" or letra == "o" or letra == "u":
  print(f"A letra {letra} é uma vogal")
else:
  print(f"A letra {letra} não é uma vogal")
```

```python
salario = float(input("Digite seu salário: "))
if salario > 1500.00:
  print(f"Seu novo salário é: {salario * 1.10:.2f}")
elif salario <= 1500.00:
  print(f"Seu novo salário é: {salario * 1.15:.2f}")
else:
  print("Valor inválido.")

```
